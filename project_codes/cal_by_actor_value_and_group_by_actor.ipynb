# 배우에 따른 영화목록 재정렬/그룹화

```python
# 굳이 이렇게 할 필요가 없었습니다.
actor1 = pd.Series(data['actor1'])
actor2 = pd.Series(data['actor2'])
actor3 = pd.Series(data['actor3'])
actor4 = pd.Series(data['actor4'])
actor5 = pd.Series(data['actor5'])
#각 actor 불러오기
```

```python
actors_all = pd.concat([actor1, actor2, actor3, actor4, actor5], ignore_index=True)
len(actors_all)
```

```python
actors_uniq = actors_all.unique()
len(actors_uniq)
# #actor전체에서 중복제거, 중복없는 배우 리스트 만들기
```

```python
#배우 목록 이용해서 배우별 영화 목록 만들기
#한번에 그룹화 해도 되지만, 이렇게 해 두면 훨씬 깔끔해서 만들어보았습니다.

movie_by_actor = pd.DataFrame()
for i in actor_list:
    col1 = data[data['actor1'] == i]
    col2 = data[data['actor2'] == i]
    col3 = data[data['actor3'] == i]
    col4 = data[data['actor4'] == i]
    col5 = data[data['actor5'] == i]
    
    df = pd.concat([col1,col2,col3,col4,col5], axis=0, ignore_index=True)
    df['출연배우'] = i
    movie_by_actor = pd.concat([movie_by_actor,df])

movie_by_actor
```

```python
#그룹화, 전국관람객 요소와 전국 스크린수 요소 배우별 총합 구하기
grouped_1 = movie_by_actor['관객수 비율'].groupby(movie_by_actor['출연배우'])
grouped_2 = movie_by_actor['스크린 점유율'].groupby(movie_by_actor['출연배우'])
grouped_3 = movie_by_actor['장르'].groupby(movie_by_actor['출연배우'])

sum1 = grouped_1.sum()
sum2 = grouped_2.sum()

df_group = pd.concat([sum1,sum2], axis = 1)

```
