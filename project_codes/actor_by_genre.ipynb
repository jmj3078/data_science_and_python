```python
import pandas as pd
import numpy as np
```

```python
movie = pd.read_csv('/Users/j-max/Documents/project_python/데이터 총합본.csv')
genre = movie.groupby(['장르']).size()
print(movie.columns)
```

```python
genre = genre.sort_values(ascending=False)
print(genre)
```

```python
genre = pd.DataFrame(genre)
genre
```

```python
act = movie['영화배우목록']
#for i in range(0, len(act)):
    #act[i] = act[i].split(sep=' ')[0:5]
for i in range(0, len(act)):   
    act[i] = act[i].split()
```

```python
act
```

```python
df_act = pd.DataFrame(index = range(0,len(act)), columns = ['actor1','actor2','actor3','actor4','actor5'])

for j in range(0,5):
    col_name = 'actor{0}'.format(j+1)    
    for i in range(0,len(act)):
        df_act[col_name][i] = act[i][j]
```

```python
df_act
```

```python
result = pd.concat([movie, df_act], axis=1) 
```

```python
result
```

```python
result.to_csv("dataset_overall.csv")
```
